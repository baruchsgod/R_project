```r
#Leyendo la información del dataset con latitudes y longitudes. 
country_info <- read.csv('Country_lat_longitude.csv')
```

```r
#Renombrando columnas para realizar el inner join de la información. 
country_info <- rename(country_info, country = name, code = country)
```

```r
#Inner Join para agregar la latitud y la longitud y poder graficar en un mapa la información
data_leaflet <- data %>% inner_join(country_info)
```

```r
data_leaflet
```

```r
#Agrupando la información para graficar.  
data_leaflet <- data_leaflet %>%
                group_by(latitude, longitude, country) %>%
                summarise(suicidios = sum(suicides_no, na.rm = TRUE))
```

```r
head(data_leaflet)
```

```r
#Ordenando la información por aquellos paises que cuentan con mas suicidios en los años analizados.
df_leaflet <- data_leaflet %>% 
                arrange(desc(suicidios))
```

```r
#Agregando un mínimo de Zoom en el mapa. 
opciones <- leafletOptions(minZoom=2)
#Agregando una paleta de colores. 
pal <- colorNumeric(palette='YlOrRd', domain=df_leaflet$suicidios)
```

```r
html <- "La cantidad de suicidios entre 1985 y 2016 es de:"
```

```r
leaflet(df_leaflet[1:20,],options = opciones) %>% addProviderTiles(providers$Stamen.Terrain) %>%
    addCircleMarkers(lng=~longitude, lat=~latitude,radius=~suicidios^0.2,color=~pal(suicidios),fillOpacity=0.8,
                     popup=~paste("<font face = 'Arial'><b>",country,"</b><br>" ,html,"<b>",
                                  format(suicidios,big.mark=",",scientific=FALSE),"</b>",
                                            "<br>","<img src='suicide_gun.webp' width='100px'></font>"))
```

```r
getting_high_years <- function(data){
    data <- data
    years <- distinct(data['year']) %>% arrange(year)
    max <- 0
    year <- 0
    country <- ""
    n <- 0
    for(i in 1:dim(years)[1]){
        for(j in 1:dim(data)[1]){
            if(data[j,2] == years[i,1]){
                if(data[j,3] > max){
                    max <- data[j,3]
                    year <- years[i,1]
                    country <- data[j,1]
                } 
            }
        }
        if(n == 0){
            df <- data.frame('year' = year,
                             'country' = country,
                             'total' = max)
            n <- 2
        }else{
            df[nrow(df) + 1,] <- c(year, country, max)
        }
        max <- 0
    }
    df
}
```

¿Qué es la distopía?

Dystopia es un país imaginario que tiene las personas menos felices del mundo. El propósito de establecer Dystopia es tener un punto de referencia contra el cual todos los países puedan compararse favorablemente (ningún país se desempeña peor que Dystopia) en términos de cada una de las seis variables clave, lo que permite que cada subbarra tenga un ancho positivo. Las puntuaciones más bajas observadas para las seis variables clave, por lo tanto, caracterizan la distopía. Dado que la vida sería muy desagradable en un país con los ingresos más bajos del mundo, la esperanza de vida más baja, la generosidad más baja, la mayor corrupción, la menor libertad y el menor apoyo social, se lo denomina "Distopía", en contraste con Utopía.

¿Qué son los residuos?

Los residuos, o componentes no explicados, difieren para cada país, lo que refleja la medida en que las seis variables sobreexplican o subexplican las evaluaciones de vida promedio de 2014-2016. Estos residuos tienen un valor promedio de aproximadamente cero en el conjunto de países. La Figura 2.2 muestra el residuo promedio para cada país cuando se aplica la ecuación de la Tabla 2.1 a los datos promedio de 2014-2016 para las seis variables en ese país. Combinamos estos residuos con la estimación de las evaluaciones de vida en Dystopia para que la barra combinada siempre tenga valores positivos. Como puede verse en la Figura 2.2, aunque algunos residuos de la evaluación de la vida son bastante grandes, en ocasiones superan un punto en la escala de 0 a 10, siempre son mucho más pequeños que el valor calculado en Dystopia, donde la vida media se valora en 1,85 en la escala de 0 a 10.
